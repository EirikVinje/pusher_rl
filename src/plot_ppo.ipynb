```python
import os
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
path = f"/home/{os.environ['USER']}/data/pusher_models/PPO_reward.csv"
path2 = f'/home/{os.environ["USER"]}/data/pusher_models/PPO_valueloss.csv'
path3 = f'/home/{os.environ["USER"]}/data/pusher_models/PPO_loss.csv'

df = pd.read_csv(path, encoding='utf-8')

#df = df.drop(df.columns[0], axis=1)
#df.to_csv(f'/home/{os.environ["USER"]}/data/pusher_models/PPO_reward.csv', index=False)

df2 = pd.read_csv(path2, encoding='utf-8')

#df2 = df2.drop(df2.columns[0], axis=1)
#df2.to_csv(f'/home/{os.environ["USER"]}/data/pusher_models/PPO_valueloss.csv', index=False)

df3 = pd.read_csv(path3, encoding='utf-8')
#df3 = df3.drop(df3.columns[0], axis=1)
#df3.to_csv(f'/home/{os.environ["USER"]}/data/pusher_models/PPO_loss.csv', index=False)

rewards = list(df["Value"])
steps = list(df["Step"])
value_loss = list(df2["Value"])
loss = list(df3["Value"])
```

```python
x = steps
y = rewards

fig, ax = plt.subplots()


ax.plot(x, y)

ax.set_xlabel("Episode")
ax.set_ylabel("Reward")
ax.set_title("Reward curve")

plt.show()
```

```python
x = steps
y = value_loss

fig, ax = plt.subplots()


ax.plot(x, y)

ax.set_xlabel("Episode")
ax.set_ylabel("Value loss")
ax.set_title("Value loss curve")
ax.set_ylim(0, 0.2)

plt.show()
```

```python
x = steps
y = loss

fig, ax = plt.subplots()


ax.plot(x, y)

ax.set_xlabel("Episode")
ax.set_ylabel("Loss")
ax.set_title("Loss curve")
ax.set_ylim(-1, 30)
ax.set_xlim(0, 200_000)

plt.show()
```
